Because stock markets are very volatile and prices changes rapidly. So to capture any trend we should smooth out data. So I will be taking 2 month and 4 month average.

```python
def MovingAvg(df, cmp):
    df['MA2M'] = df['open'].rolling(60).mean()
    df['MA4M'] = df['open'].rolling(120).mean()
    df['close'].plot(figsize = (18,9))
    df['MA2M'].plot()
    df['MA4M'].plot()
    plt.title('Moving avg of '+cmp+' stocks')
    plt.legend()
```

```python
MovingAvg(goog, 'Alphabet')
```

```python
MovingAvg(apl, 'Apple')
```

```python
MovingAvg(amz, 'Amazon')
```

Because all three are IT companies. So maybe there is some relation b/w stocks of these companies. Let's see if there is some correlation b/w these three stocks.

```python
tech_gnt = pd.concat([goog['open'], apl['open'], amz['open']], axis=1)
tech_gnt.columns = ['Alphabet Open', 'Apple Open', 'Amazon Open']
```

```python
g = sns.pairplot(tech_gnt, diag_kws = {'bins':40}, plot_kws = {'s':15})
g.fig.set_size_inches(12,9)
```

As we can see from the plot above that there is somewhat positive linear correlation between opening stocks prizes of alphabet and amazon.

## Daily Percentage Gain

$$r_t = {p_t \over p_{t-1}}-1$$


```python
goog['returns'] = (goog['close']/goog['close'].shift(1))-1
apl['returns'] = (apl['close']/apl['close'].shift(1))-1
amz['returns'] = (amz['close']/amz['close'].shift(1))-1
```

```python
plt.hist(goog['returns'], alpha=0.5, label='Alphabet Stocks Returns', bins=50, color='b')
plt.hist(apl['returns'], alpha=0.5, label='Apple Stocks Returns', bins=50, color='r')
plt.hist(amz['returns'], alpha=0.5, label='Amazon Stocks Returns', bins=50, color='g')
plt.legend()
matplotlib.rcParams['figure.figsize'] = (12, 9)
```

```python
goog['returns'].plot(kind='kde',label='Alphabet Stocks Returns', color='b')
apl['returns'].plot(kind='kde', label='Apple Stocks Returns', color='r')
amz['returns'].plot(kind='kde',label='Amazon Stocks Returns', color='g')
plt.legend()
```

So looking at the plots we can say that stocks of alphabet are a little less volatile then other two. Although all three are similar in term of stability

```python
gnt_returns = pd.concat([goog['returns'], apl['returns'], amz['returns']], axis=1)
gnt_returns.columns = ['Alphabet Returns', 'Apple Returns', 'Amazon Returns']

sns.boxplot(data=gnt_returns)
```

```python
sns.pairplot(gnt_returns)
```

PairPlot b/w returns also shows there is some kind of relation b/w amazon and alphabet stocks.

## Cumulative Return

$$i_t = (1+r_t)i_{t-1}$$

```python
goog['Cumulative Return'] = (1+goog['returns']).cumprod()-1
apl['Cumulative Return'] = (1+apl['returns']).cumprod()-1
amz['Cumulative Return'] = (1+amz['returns']).cumprod()-1
```

```python
goog['Cumulative Return'].plot(label = "Alphabet", figsize = (18, 9))
apl['Cumulative Return'].plot(label = 'Apple')
amz['Cumulative Return'].plot(label = 'Amazon')
plt.legend()
```

Cumulative returns os alphabet is lower than the other two. Most of the times return of apple's stocks are highest. But in recent time returns on amazon's stcoks have drastically increased.
